```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data=pd.read_csv('Diamonds_Prices2022.csv')
data.shape
```

```python
size=(20,11)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)
```

```python
sns.countplot(x='cut',data=data)
plt.xticks(rotation=90,fontsize=18,color='k')
plt.yticks(color='k',fontsize=18)
plt.xlabel("Cut Quality",fontsize=36)
plt.ylabel("Price",fontsize=36)
plt.show()
```

```python
sns.barplot(x='cut',y='price',data=data)
plt.xticks(rotation=90,fontsize=18,color='k')
plt.yticks(color='k',fontsize=18)
plt.xlabel("Cut Quality",fontsize=36)
plt.ylabel("Price",fontsize=36)
plt.show()
```

```python
size=(10,4)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)

plt.subplot(1,2,1)
sns.boxplot(x='color',y='depth',data=data)
plt.xlabel('Diamond Color',fontsize=14,fontweight='bold',color='r')
plt.ylabel('Diamond Depth',fontsize=14,fontweight='bold',color='b')
plt.title('Box Plot')

plt.subplot(1,2,2)
sns.boxplot(x='depth',y='color',data=data,orient='h')
plt.xlabel('Diamond Depth',fontsize=10,fontweight='bold',color='b')
plt.ylabel('Diamond Color',fontsize=10,fontweight='bold',color='r')
plt.title('Boxplot With Orient')

plt.show()
```

```python
size=(10,4)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)

sns.lineplot(x='clarity',y='price',data=data)
plt.xlabel('Diamond Clarity',fontsize=10,fontweight='bold',color='b')
plt.ylabel('Price',fontsize=10,fontweight='bold',color='r')
plt.title('Line Plot')
plt.show()
```

```python
x='cut'
y='price'
print(len(x))
col=['red','blue','cyan','purple','yellow']
myexplode=[0,0,0,0,0]
plt.pie(y,labels=x,data=data,colors=col,startangle=0,explode=myexplode,shadow=True)
plt.title('Pie Chart')
plt.show()
```

```python
size=(10,5)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)

sns.scatterplot(x='carat',y='price',data=data)
plt.xlabel('carat',fontsize=10,fontweight='bold',color='b')
plt.ylabel('Price',fontsize=10,fontweight='bold',color='r')
plt.title('Scatter Plot')
plt.show()
```

```python
x=data.pivot_table(index='carat',columns='cut',values='price')
size=(10,8)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)
sns.heatmap(x)
plt.xlabel('Diamond Cut Quality',fontsize=18)
plt.ylabel('Carat',fontsize=18)
plt.title('Heat Map Showing the Price of Diamond as per  Cut quality and Carat Size')
plt.show()
```

```python
sns.jointplot(x='carat',y='depth',data=data)
plt.xlabel('Carats',fontsize=18)
plt.ylabel('Depth',fontsize=18)
plt.title('Joint plot')
plt.show()
```

```python
size=(20,11)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)
sns.violinplot(x='clarity',y='depth',data=data,hue='cut')
plt.xlabel('Diamond Clarity',fontsize=28,color='g')
plt.ylabel('Depth',fontsize=28,color='r')
plt.title('Violinplot',fontsize=36)
plt.show()
```

```python
size=(20,11)
fig,ax=plt.subplots(figsize=size,constrained_layout=True)
sns.stripplot(x='cut',y='price',data=data,hue='cut')
plt.show()
```
